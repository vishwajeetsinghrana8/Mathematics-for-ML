## What is Linear Algebra?

Linear algebra is a branch of mathematics that deals with **vectors**, **matrices**, and **linear transformations**. It is a fundamental tool in many areas of science and engineering, including computer graphics, machine learning, and data analysis.

### Core Concepts

1.  **Vectors**:
    *   A vector is a mathematical object that has both **magnitude** (or length) and **direction**.
    *   It can be represented as an arrow in a coordinate system or as a list of numbers (components).
    *   For example, in a 2D plane, a vector can be written as `[x, y]`.

2.  **Matrices**:
    *   A matrix is a rectangular array of numbers, symbols, or expressions, arranged in rows and columns.
    *   Matrices are used to represent and solve systems of linear equations, and to perform linear transformations.

3.  **Linear Transformations**:
    *   A linear transformation is a function between two vector spaces that preserves the operations of vector addition and scalar multiplication.
    *   In simpler terms, it's a way of moving, rotating, or scaling vectors in a space.

### Why is Linear Algebra Important?

*   **Solving Systems of Equations**: Linear algebra provides powerful methods for solving systems of linear equations, which appear in countless applications.
*   **Computer Graphics**: It's used for everything from rendering 3D scenes to applying special effects.
*   **Machine Learning**: Many machine learning algorithms, such as linear regression and principal component analysis (PCA), are based on linear algebra.
*   **Data Representation**: Data can often be represented as vectors or matrices, making it easy to manipulate and analyze using linear algebra techniques.

## Plot an Example Linear Equation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 50, 1000)
```

```python
x = 2.5 * t
```

```python
y = 4 * (t - 6)
```

```python
fig, ax = plt.subplots()
plt.title("Meeting point of x and y")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0, 50])
ax.set_ylim([0, 100])
ax.plot(t, x, c="green")
ax.plot(t, y, c="brown")
plt.axvline(x=16, color='black', linestyle='--')
_ = plt.axhline(y=40, color='black', linestyle='--')
```

## **What are Tensors ?**

A **tensor** is a mathematical object that generalizes scalars, vectors, and matrices to higher dimensions. In simple terms, a tensor is a **multi-dimensional array** of numbers that can be used to represent data in various dimensions.

### Types of Tensors

| Tensor Type | Order (Rank) | Example                    |
| ----------- | ------------ | -------------------------- |
| Scalar      | 0            | `5` (just a single number) |
| Vector      | 1            | `[3, 7, 2]`                |
| Matrix      | 2            | `[[1, 2], [3, 4]]`         |
| 3D Tensor   | 3            | `[[[1], [2]], [[3], [4]]]` |
| n-D Tensor  | n            | Multi-dimensional arrays   |

### Visualization of Tensors

* **0D**: Point (single number)
* **1D**: Line (vector)
* **2D**: Table (matrix)
* **3D**: Cube (like an RGB image with depth)
* **4D+**: Complex data structures used in deep learning

## **Scalar Tensors in Tensorflow**

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
scalar_tf_x = tf.Variable(10)
scalar_tf_y = tf.Variable(15)
print(scalar_tf_x)
print(scalar_tf_y)
```

```python
scalar_tf_x = tf.Variable(10, dtype=tf.int16)
scalar_tf_y = tf.Variable(15, dtype=tf.int16)
print(scalar_tf_x)
print(scalar_tf_y)
```

```python
scalar_tf_x.shape
```

```python
scalar_tf_x + scalar_tf_y
```

```python
tf_tensor_sum = tf.add(scalar_tf_x, scalar_tf_y)
tf_tensor_sum
```

```python
## You can also convert them into NumPy array
tf_tensor_sum.numpy()
```

```python
type(tf_tensor_sum.numpy())
```

```python
## Just like that you can also create you float tensors
tf_tensor_float1 = tf.Variable(20.0, dtype=tf.float16)
tf_tensor_float2 = tf.Variable(10.0, dtype=tf.float16)
print(tf_tensor_float1)
print(tf_tensor_float2)
```

```python
tf_tensor_sum_float = tf.add(tf_tensor_float1, tf_tensor_float2)
tf_tensor_sum_float
```

## What are Vector Tensors?

A **vector** is a tensor with one dimension. It's an ordered list of numbers. In linear algebra, vectors are used to represent points in space, directions, and magnitudes. They are also known as **1-dimensional tensors** or **rank-1 tensors**.
